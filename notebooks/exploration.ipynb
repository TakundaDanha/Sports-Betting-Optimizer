```python


import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load bankroll progression data
df = pd.read_csv("../data/processed/bankroll_progression.csv")

plt.figure(figsize=(12, 6))
plt.plot(df.index, df["flat_betting"], label="Flat Betting")
plt.plot(df.index, df["kelly_criterion"], label="Kelly Criterion")
plt.title("Bankroll Progression Over Time")
plt.xlabel("Number of Bets")
plt.ylabel("Bankroll ($)")
plt.legend()
plt.grid(True)
plt.show()

# Summary stats
def max_drawdown(series):
    roll_max = series.cummax()
    drawdowns = (series - roll_max) / roll_max
    return drawdowns.min()

flat_final = df["flat_betting"].iloc[-1]
kelly_final = df["kelly_criterion"].iloc[-1]

flat_drawdown = max_drawdown(df["flat_betting"])
kelly_drawdown = max_drawdown(df["kelly_criterion"])

print(f"Flat Betting Final Bankroll: ${flat_final:.2f}")
print(f"Kelly Criterion Final Bankroll: ${kelly_final:.2f}")
print(f"Flat Betting Max Drawdown: {flat_drawdown:.2%}")
print(f"Kelly Criterion Max Drawdown: {kelly_drawdown:.2%}")

# Optional: Calculate total return percentage
flat_return = (flat_final - df["flat_betting"].iloc[0]) / df["flat_betting"].iloc[0]
kelly_return = (kelly_final - df["kelly_criterion"].iloc[0]) / df["kelly_criterion"].iloc[0]

print(f"Flat Betting Total Return: {flat_return:.2%}")
print(f"Kelly Criterion Total Return: {kelly_return:.2%}")
```
